```python
import matplotlib.pyplot as plt

import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import numpy

np.set_printoptions(suppress=True)
%matplotlib inline

from sklearn.preprocessing import MinMaxScaler
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
import seaborn as sns 
sns.set_palette("husl") #设置所有图的颜色，使用hls色彩空间
import numpy as np
from sklearn.metrics import f1_score #均方误差
from sklearn.metrics import classification_report #平方绝对误差

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
```

```python
#读取 cleandata附件1.csv
data1=pd.read_csv('cleandata附件1.csv',encoding='gbk')
data1.head()
```

```python
#读取 cleandata附件2.csv
data2=pd.read_csv('cleandata附件2.csv',encoding='gbk')
data2.head()
```

```python
data1.columns
```

```python
data2.columns
```

```python
data=pd.concat([data1,data2])
data.shape
```

```python
data.isnull().sum()
```

```python
data["('金额', 'std')_x"].fillna(data["('金额', 'std')_x"].mean(),inplace=True)
data["('税额', 'std')_x"].fillna(data["('税额', 'std')_x"].mean(),inplace=True)
data["('价税合计', 'std')_x"].fillna(data["('价税合计', 'std')_x"].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data['是否违约']=data['是否违约'].map({'是':1,"否":0})
data["('发票状态', 'top')"]=data["('发票状态', 'top')"].map({'有效发票':1,"作废发票":0})
del data["('发票状态', 'top')_x"]
```

```python
train_X=data[data['是否违约'].isnull()==False][data.columns[3:]]
train_Y=data[data['是否违约'].isnull()==False]['是否违约']
pred_X=data[data['是否违约'].isnull()==True][data.columns[3:]]
```

```python
from collections import Counter
Counter(train_Y)
```

```python
from imblearn.over_sampling import SMOTE
X_resampled, y_resampled = SMOTE().fit_resample(train_X, train_Y)
print(sorted(Counter(y_resampled).items()))
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y = train_test_split(X_resampled,y_resampled,test_size=0.3,random_state=5)
```

```python
model = RandomForestClassifier()
model.fit(train_X,train_y)

print(classification_report(test_y, model.predict(test_X)))
```

```python
model = AdaBoostClassifier()
model.fit(train_X,train_y)

print(classification_report(test_y, model.predict(test_X)))
```

```python
model = SVC()
model.fit(train_X,train_y)

print(classification_report(test_y, model.predict(test_X)))
```

```python
from xgboost.sklearn import XGBClassifier

model = XGBClassifier()
model.fit(train_X,train_y)

print(classification_report(test_y, model.predict(test_X)))
```
